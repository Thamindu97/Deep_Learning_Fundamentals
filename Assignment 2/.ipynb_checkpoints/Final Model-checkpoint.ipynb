## Data Loading & Pre-processing

* Total number of images: 90483
* Training set size: 67692 images (one fruit or vegetable per image)
* Test set size: 22688 images (one fruit or vegetable per image)
* Number of classes: 131 (fruits and vegetables)

```python
# Setup paths to the training and test directories
training_dir = 'fruits_dataset/fruits-360_dataset/fruits-360/Training'
test_dir = 'fruits_dataset/fruits-360_dataset/fruits-360/Test'

# Using ImageDataGenerator for data augmentation and normalization
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest',
    validation_split=0.1  # using 10% of the training data for validation
)

test_datagen = ImageDataGenerator(rescale=1./255)

# Load images from the directories and dataset split
train_generator = train_datagen.flow_from_directory(
    training_dir,
    target_size=(100, 100),
    batch_size=32,
    class_mode='categorical',
    subset='training'  # set as training data
)

validation_generator = train_datagen.flow_from_directory(
    training_dir,  # same directory as training data
    target_size=(100, 100),
    batch_size=32,
    class_mode='categorical',
    subset='validation'  # set as validation data
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(100, 100),
    batch_size=32,
    class_mode='categorical'
)
```

The ImageDataGenerator class in Keras is used to augment your image data in real-time while your model is training. It generates batches of tensor image data with real-time data augmentation, effectively increasing the diversity of your training set without the need to explicitly store additional images. This can help prevent overfitting and can lead to better generalization of your model.

* rescale: Rescales the image by a factor provided (in this case, 1./255). Since image data can range from 0 to 255, this rescale factor is used to scale the images to a range of 0 to 1, which is a common practice in image processing that helps with the training process.

* rotation_range: Specifies the degree range within which to randomly rotate pictures. In this case, images will be rotated by a random amount in the range [-20, 20] degrees.

* width_shift_range: Fraction of total width, if < 1, or pixels if >= 1. This is the range within which to randomly translate pictures horizontally. A value of 0.2 means that the image can be translated by up to +/- 20% of its width.

* height_shift_range: Same as width_shift_range, but for vertical translation.

* shear_range: Shear Intensity (Shear angle in counter-clockwise direction as radians). It allows the image to be sheared by a specified angle, creating a sort of "stretching" effect.

* zoom_range: Range for random zoom. If a float, [lower, upper] = [1-zoom_range, 1+zoom_range]. For zoom_range=0.2, the image can be randomly zoomed in or out by 20%.

* horizontal_flip: Boolean value. If True, randomly flips inputs horizontally, which can be useful for natural images as it's a realistic form of augmentation.

* fill_mode: Points outside the boundaries of the input are filled according to the given mode, with 'nearest' being the default (and typically the preferred) strategy. It fills in new pixels that can appear after a rotation or a width/height shift.

* validation_split: Float between 0 and 1. The fraction of the data to reserve for validation. In this case, 10% of the data will be used as the validation set. When this parameter is set, the flow_from_directory method will allow you to split the data into training and validation sets by using the subset argument.

```python
# Function to count images in each subdirectory
def count_images(directory):
    counts = {}
    for class_folder in os.listdir(directory):
        class_folder_path = os.path.join(directory, class_folder)
        if os.path.isdir(class_folder_path):
            counts[class_folder] = len(os.listdir(class_folder_path))
    return counts

# Count images in each dataset
train_counts = count_images(training_dir)
test_counts = count_images(test_dir)

# Assuming a validation split of 0.1 from the training data
validation_split = 0.1
validation_counts = {k: int(v * validation_split) for k, v in train_counts.items()}
train_counts = {k: v - validation_counts[k] for k, v in train_counts.items()}

# Create a DataFrame for visualization
data = {
    'Class': list(train_counts.keys()),
    'Training': list(train_counts.values()),
    'Validation': list(validation_counts.values()),
    'Test': list(test_counts.values())
}
df = pd.DataFrame(data)
df_melted = df.melt(id_vars='Class', var_name='Dataset', value_name='Number of Images')

plt.figure(figsize=(18, 8))

# Create a barplot
sns.barplot(x='Class', y='Number of Images', hue='Dataset', data=df_melted , palette='viridis')

# Improve the legibility of the plot
plt.xticks(rotation=90)
plt.title('Distribution of Images per Class in Training, Validation, and Test Sets')
plt.xlabel('Class')
plt.ylabel('Number of Images')

# Show the plot
plt.tight_layout()
plt.show()

# Since all images are 100x100, we don't need to visualize this.
print(f"Image resolution is fixed at 100x100 pixels.")
```

```python
# Function to count classes in each dataset directory
def count_classes(directory):
    return len(os.listdir(directory))

# Count classes in each dataset
num_classes_train = count_classes(training_dir)
num_classes_test = count_classes(test_dir)

# Assuming the same classes are present in each dataset and that the validation split is 10%
num_classes_validation = num_classes_train  # same as training

# Data for visualization
data = {
    'Dataset': ['Training', 'Validation', 'Test'],
    'Number of Classes': [num_classes_train, num_classes_validation, num_classes_test]
}

# Convert to DataFrame
df_classes = pd.DataFrame(data)

# Visualization
plt.figure(figsize=(8, 6))
sns.barplot(x='Dataset', y='Number of Classes', data=df_classes, palette='viridis')
plt.title('Total Number of Classes in Each Dataset')
plt.xlabel('Dataset')
plt.ylabel('Number of Classes')
plt.ylim(0, max(data['Number of Classes']) + 5)  # Adjust y-axis limit for better visualization
plt.show()
```

## Model Implementation 

### DenseNet with  pre-trained weights

```python
# Load the DenseNet121 network with pre-trained weights from ImageNet
DenseNet_base_model = DenseNet121(weights='imagenet', include_top=False, input_tensor=Input(shape=(100, 100, 3)))

# Freeze the layers in the base model
for layer in DenseNet_base_model.layers:
    layer.trainable = False

# Construct the head of the model to be placed on top of the base model
x = DenseNet_base_model.output
x = GlobalAveragePooling2D()(x)
x = BatchNormalization()(x)
x = Dropout(0.5)(x)
x = Dense(256, activation='relu')(x)
x = BatchNormalization()(x)
x = Dropout(0.5)(x)
head_model = Dense(num_classes, activation='softmax')(x)  # num_classes should be set to the number of your classes

# Create the final model
DenseNet_model = Model(inputs=DenseNet_base_model.input, outputs=head_model)

# Compile the model
DenseNet_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# View the structure of the model
DenseNet_model.summary()
```

This code does the following:

1. Loads the DenseNet121 model with pre-trained ImageNet weights, excluding the top (head) of the network which is responsible for the classification.
2. Freezes the pre-trained layers to prevent them from being updated during the first phase of training.
3. Adds a new head to the network which includes global average pooling and fully connected layers tailored to the number of classes in your dataset.
4. Compiles the model with the Adam optimizer and categorical crossentropy loss function, which is appropriate for multi-class classification tasks.

## Model Optimization and Tuning

### DenseNet with  pre-trained weights

```python
def build_model(hp):
    # Load the DenseNet121 network with pre-trained weights from ImageNet
    base_model = DenseNet121(weights='imagenet', include_top=False, input_tensor=Input(shape=(100, 100, 3)))
    base_model.trainable = False  # Freeze the layers in the base model

    # Start building the model
    model = Sequential()
    model.add(base_model)
    model.add(GlobalAveragePooling2D())
    model.add(BatchNormalization())
    model.add(Dropout(0.5))

    # Use hyperparameter for activation function
    activation_choice = hp.Choice('activation', values=['relu', 'sigmoid'])

    model.add(Dense(units=256, activation=activation_choice))
    model.add(BatchNormalization())
    model.add(Dropout(0.5))
    model.add(Dense(num_classes, activation='softmax'))  # Assuming num_classes is set correctly

    # Use hyperparameter for learning rate
    learning_rate_choice = hp.Choice('learning_rate', values=[0.1, 0.01, 0.001, 0.0001])

    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate_choice),
                  loss='categorical_crossentropy',
                  metrics=['accuracy'])

    return model

# Initialize the tuner
tuner = kt.Hyperband(
    build_model,
    objective='val_accuracy',
    max_epochs=20,
    hyperband_iterations=2,  # Number of times to iterate over the entire Hyperband algorithm
    directory='keras_tuner_dir',
    project_name='densenet_tuning'
)

# Define the callbacks list
callbacks_list = [
    tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True),
    tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5, min_lr=1e-5),
    tf.keras.callbacks.ModelCheckpoint(filepath='best_model.h5', monitor='val_accuracy', save_best_only=True)
]

# Perform the hyperparameter tuning
tuner.search(
    train_generator,
    steps_per_epoch=train_generator.samples // 32,  # Here we use 64 for batch size, but this will be tuned so it's just a placeholder
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // 32,  # Same for validation steps
    epochs=20,
    callbacks=callbacks_list
)

# Get the best hyperparameters
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
print(f"""
The hyperparameter search is complete. The optimal learning rate is {best_hps.get('learning_rate')}
and the optimal activation function is {best_hps.get('activation')}.
""")

```

```python
# Build the model with the best hyperparameters and train it on the data for 50 epochs
best_model = tuner.hypermodel.build(best_hps)

best_model.save('Dense_best_model.h5')  # Saves the model to an HDF5 file
```

```python
# Load the model back from the HDF5 file
best_model = keras.models.load_model('Dense_best_model.h5')
```

```python

# Build the model with the best hyperparameters and train it on the data for 50 epochs
#best_model = tuner.hypermodel.build(best_hps)

best_model_history = best_model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // 32,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // 32,
    epochs=50,
    callbacks=callbacks_list
)
```

### Predictions

```python


def plot_images_with_predictions(test_generator, model, num_images=10):
    # Get a batch of images from the test generator
    test_images, test_labels_one_hot = next(test_generator)
    
    # Get predictions
    pred_probs = model.predict(test_images)
    pred_classes = np.argmax(pred_probs, axis=1)
    
    # Convert one-hot encoded labels to indices
    test_labels = np.argmax(test_labels_one_hot, axis=1)
    
    # Get the corresponding class names
    label_map = dict((v,k) for k,v in test_generator.class_indices.items())
    true_class_names = [label_map[label] for label in test_labels]
    pred_class_names = [label_map[pred] for pred in pred_classes]
    
    # Plot the images with predictions
    plt.figure(figsize=(15, 15))
    for i in range(num_images):
        plt.subplot(1, num_images, i+1)
        plt.imshow(test_images[i])
        plt.title(f"True: {true_class_names[i]}\nPred: {pred_class_names[i]}")
        plt.axis('off')
    plt.show()

# Call the function to plot the images
plot_images_with_predictions(test_generator, best_model, num_images=5)

```
